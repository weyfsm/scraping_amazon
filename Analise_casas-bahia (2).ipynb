# Preço Médio

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tabela =pd.read_csv('./datasetbahia1048.csv',sep=';')
tabela
```

```python
print('Quantidade de linhas e colunas:', tabela.shape)
print('Quantidade de dados nulos:', tabela.isna().sum().sum())
```

```python
tabela['preco']=tabela['preco'].astype(float)
```

```python
tabela.dtypes
```

```python
tabela.describe()
```

```python
casas_b = tabela[['nome','preco']]
casas_b.head(50)
```

```python
data=casas_b[casas_b['nome'].str.contains("Notebook")]
datalenovo=data[data['nome'].str.contains("Lenovo")]
datasamsung=data[data['nome'].str.contains("Samsung")]
dataacer=data[data['nome'].str.contains("Acer")]
datadell=data[data['nome'].str.contains("Dell")]
datapositivo=data[data['nome'].str.contains("Positivo")]
datamultilaser=data[data['nome'].str.contains("Multilaser")]
datahp=data[data['nome'].str.contains("Hp")]
datavaio=data[data['nome'].str.contains("Vaio")]
datalg=data[data['nome'].str.contains("LG")]
dataasus=data[data['nome'].str.contains("ASUS")]
data.to_csv('./datasetbahia.csv',sep=';',index=False,encoding=('utf-8-sig'))
```

```python
marcas=['GERAL','Lenovo','Samsung','Acer','Dell','Positivo','Multilaser', 'Hp','Vaio','LG','ASUS']
mediamarcas=[float(data.mean()),float(datalenovo.mean()),float(datasamsung.mean()),float(dataacer.mean()),float(datadell.mean()),float(datapositivo.mean()), float(datamultilaser.mean()),float(datahp.mean()),float(datavaio.mean()),float(datalg.mean()),float(dataasus.mean())]
quantidademarca=[len(data),len(datalenovo),len(datasamsung),len(dataacer),len(datadell),len(datapositivo),len(datamultilaser),len(datahp),len(datavaio),len(datalg),len(dataasus)]
dictmarcas = {'Marcas':marcas, 'Preço Médio':mediamarcas,'Qtd. de Produtos':quantidademarca}
medias=pd.DataFrame(dictmarcas).round(2)
medias.to_csv('./medias.csv',sep=';',index=False,encoding=('utf-8-sig'))
medias
```

```python

plt.figure(figsize = (12, 5))
plt.bar(marcas, mediamarcas, color='blue')
plt.title("Preço Médio por Marca",fontsize=18)
plt.xlabel("Marcas de Notebooks", fontsize= 14)
plt.ylabel("Média de Preços (R$)",fontsize= 14)

plt.savefig('media_Casas_bahia.png')
```

# Avaliação 

```python
aval = tabela[['nome','avali']]
```

```python
#Seleção dos notebooks com avaliação igual a 5:
avali = aval.sort_values(by='avali', ascending = False)
melhores = avali.head(111)
melhores
```

```python
data2=melhores[melhores['nome'].str.contains("Notebook")]
datalenovo2=data2[data2['nome'].str.contains("Lenovo")]
datasamsung2=data2[data2['nome'].str.contains("Samsung")]
dataacer2=data2[data2['nome'].str.contains("Acer")]
datadell2=data2[data2['nome'].str.contains("Dell")]
datapositivo2=data2[data2['nome'].str.contains("Positivo")]
datamultilaser2=data2[data2['nome'].str.contains("Multilaser")]
datahp2=data2[data2['nome'].str.contains("Hp")]
datavaio2=data2[data2['nome'].str.contains("Vaio")]
datalg2=data2[data2['nome'].str.contains("LG")]
dataasus2=data2[data2['nome'].str.contains("ASUS")]
data2.to_csv('./datasetbahia2.csv',sep=';',index=False,encoding=('utf-8-sig'))
```

```python
marcas2=['Lenovo','Samsung','Acer','Dell','Positivo','Multilaser', 'Hp','Vaio','LG','ASUS']
quantidademarca2=[len(datalenovo2),len(datasamsung2),len(dataacer2),len(datadell2),len(datapositivo2),len(datamultilaser2),len(datahp2),len(datavaio2),len(datalg2),len(dataasus2)]
dictmarcas2 = {'Marcas':marcas2,'Qtd. de Produtos':quantidademarca2}
melhores2=pd.DataFrame(dictmarcas2)
melhores2.to_csv('./melhores2.csv',sep=';',index=False,encoding=('utf-8-sig'))
melhores2
```

```python
plt.figure(figsize = (12, 5))
plt.bar(marcas2,quantidademarca2 , color='red')
plt.title("Marcas com melhores avaliações",fontsize=18)
plt.xlabel("Marcas de Notebooks", fontsize= 14)
plt.ylabel("Qt. de Produtos",fontsize= 14)
plt.savefig('avaliacao.png')
```
